```python
def get_user_question(row):
    
    title = row.essay_title
    text = row.essay_text
    acs_l = ast.literal_eval(row.argument_components)

    #essay = f"""### Essay title: {title}\n\n""" + f"""### Essay text:\n{text}\n\n"""
    essay = f"""### Essay text:\n""" + f"""{title}""" + "\n\n" + f"""{text}\n\n"""
    #acs = "### List of arguments in the essay: \n" + '\n'.join(f"{i + 1}. {ac}" for i, ac in enumerate(acs_l)) + "\n\n"
    
    return essay# + acs
```

```python
def get_output(row):
    
    labels_d = {'MajorClaim': "M", 'Claim': "C", 'Premise': "P"}
    #rel_d = {'supports': "S", 'attacks': "A", 'For': "S", 'Against': "A"}
    
    labels_l = ast.literal_eval(row.argument_types)
    arg_start = [int(elem) for elem in ast.literal_eval(row.arg_start_bounds)]
    arg_end = [int(elem) for elem in ast.literal_eval(row.arg_end_bounds)]
    
    sorted_indices = sorted(range(len(arg_start)), key=lambda i: arg_start[i])
    
    arg_start = [arg_start[i] for i in sorted_indices]
    arg_end = [arg_end[i] for i in sorted_indices]
    labels = [labels_l[i] for i in sorted_indices]
    
    #print(arg_start)
    
    labels_l = [labels_d[v] for v in labels]
    
    labels_l = [(arg_s, arg_e, l) for arg_s, arg_e, l in zip(arg_start, arg_end, labels_l)]
    
    output_indic = "### Output:\n"
    ac_types = f'{{"arguments": {labels_l}}}\n'
    
    
    
    # stance_pairs = ast.literal_eval(row.stance_pairs)
    # rel_types = ast.literal_eval(row.relation_types) + [pair[2] for pair in stance_pairs]   
    
    
    
    # rel_sources = [elem.replace("T", "") for elem in ast.literal_eval(row.source_args)] + [pair[0] for pair in stance_pairs]
    # rel_targets = [elem.replace("T", "") for elem in ast.literal_eval(row.target_args)]  + [pair[1] for pair in stance_pairs]
    # rels_l = [(int(s), int(t)) for s,t in zip(rel_sources, rel_targets)]
    
    # labels_l = [labels_d[v] for v in labels_l]
    # rel_types = [rel_d[v] for v in rel_types]
    
    # output_indic = "### Output:\n"
    # ac_types = f'{{"argument_types": {labels_l}}}\n'
    #rels_types = f'{{"relation_types": {[(s, t, rel_type) for (s, t), rel_type in zip(rels_l, rel_types)]}}}\n'
    
    #rels = f'{{"related_arguments": {rels_l}}}\n'
    #rel_types = f'{{"relation_types": {rels_types}}}\n'
    
    #return output_indic + ac_types + rels + rel_types
    return output_indic + ac_types# + rels_types
```

### Build prompts

```python
def formatting_fct(instruction="", input="", output=""):
    
    prompt_d ={
            
        "instruction": f"""{instruction}""",
        "input": f"""{input}""",
        "output": f"""{output}"""
            
        }
    
    return prompt_d
```

### Build data files

```python
df_train = pe_df[pe_df.split == 'TRAIN'].reset_index()

data_file_train = []

for _, row in df_train.iterrows():

    question = row.usr_text
    answer = row.output
    
    data_file_train.append(formatting_fct(instruction, question, answer))
```
