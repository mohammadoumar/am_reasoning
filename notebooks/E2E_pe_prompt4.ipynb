```python
def get_user_question(row):
    
    title = row.essay_title
    text = row.essay_text
    acs_l = ast.literal_eval(row.sorted_acs)
    
    paras_structure = ast.literal_eval(row.paras_structure)

    essay = f"""### Essay title: {title}\n\n""" + f"""### Essay text:\n{text}\n\n"""
    acs = "### List of arguments in the essay: \n" + '\n'.join(f"{i + 1}. {ac}" for i, ac in enumerate(acs_l)) + "\n\n"
    
    #paras_structure = f"""The distribution of arguments in paragraphs is as follows: {(idx, elem) for idx, elem in enumerate(paras_structure)}"""
    
    paras_structure = "The distribution of arguments in paragraphs is as follows:\n" + \
                  "\n".join([f"Paragraph {idx+1}: {elem}" for idx, elem in enumerate(paras_structure)])

    
    return essay + acs + paras_structure
```

```python
def get_output(row):
    
    labels_d = {'MajorClaim': "M", 'Claim': "C", 'Premise': "P"}
    rel_d = {'supports': "S", 'attacks': "A"}
    
    labels_l = eval(row.sorted_ac_types)
    #labels_l = [elem for sl in row.sorted_ac_types for elem in sl]
    #print(labels_l)
    
    stance_pairs = eval(row.para_ac_stances)
    stance_pairs = list({src: rel_d[relation] for src, _, relation in stance_pairs}.values())
    
    
    #print(stance_pairs)
    rel_types = eval(row.para_ac_rels)
    #print(rel_types)
    #rel_types = [elem for sl in rel_types for elem in sl if elem[2] != "NR"]# + stance_pairs
    rel_types_l = []
    
    for l in rel_types:
        rels = []
        for rel in l:
            if rel[2] != "NR":
                rels.append(rel)
        rel_types_l.append(rels)
        
    rel_types_l = [
    [(a, b, rel_d.get(rel, rel)) for (a, b, rel) in sublist]
    for sublist in rel_types_l
]
        
    #print(rel_types_l)
    #rel_types = ast.literal_eval(row.relation_types) + [pair[2] for pair in stance_pairs]   
    
    
    
    #rel_sources = [elem.replace("T", "") for elem in ast.literal_eval(row.source_args)] + [pair[0] for pair in stance_pairs]
    #rel_targets = [elem.replace("T", "") for elem in ast.literal_eval(row.target_args)]  + [pair[1] for pair in stance_pairs]
    #rels_l = [(int(s), int(t)) for s,t in zip(rel_sources, rel_targets)]
    
    labels_l = [labels_d[v] for v in labels_l]
    #rel_types_final = [(s, t, rel_d[rel_type]) for s, t, rel_type in  rel_types_l]
    #rel_types_final = ", ".join([f"Paragraph {idx+1}: {elem}" for idx, elem in enumerate(rel_types_l)])
    rel_types_final = {f"Paragraph {idx+1}": elem for idx, elem in enumerate(rel_types_l)}
    
    #print(rel_types_final)
    
    output_indic = "### Output:\n"
    ac_types = f'{{"argument_types": {labels_l}}}\n'
    stance_types = f'{{"claim_stances": {stance_pairs}}}\n'
    rels_types = f'{{"relation_types": {rel_types_final}}}\n'
    
    #rels = f'{{"related_arguments": {rels_l}}}\n'
    #rel_types = f'{{"relation_types": {rels_types}}}\n'
    
    #return output_indic + ac_types + rels + rel_types
    return output_indic + ac_types + stance_types + rels_types
```

### Build prompts

```python
def formatting_fct(instruction="", input="", output=""):
    
    prompt_d ={
            
        "instruction": f"""{instruction}""",
        "input": f"""{input}""",
        "output": f"""{output}"""
            
        }
    
    return prompt_d
```

```python
instruction = instruction = """### You are an expert in Argument Mining tasked with analyzing argumentative structures in essays. 

INPUT:
You will receive:
- An essay title.
- The complete essay text.
- An enumerated list of identified arguments extracted from the essay.

TASK 1: Argument Classification
- Classify each argument in the essay into one of the following categories: "MajorClaim"(M), "Claim"(C) or "Premise"(P).
- MajorClaim (M): The main stance or position that the author wants to prove in the essay. Usually appears in the introduction and/or conclusion.
- Claim (C): A statement that directly supports or attacks the major claim. Claims are controversial assertions that require further evidence.
- Premise (P): A reason or evidence that directly supports or attacks a claim or another premise.
- You must return a list of argument types in following JSON format: {"argument_types": [argument_types (str), argument_types (str), ..., argument_types (str)]}

TASK 2: Claim Stance Classification
- Classify each claim as either supporting (S) or attacking (A) the major claims.
- You must return a list of claim stances in following JSON format: {"claim_stances": [claim_stance (str), claim_stance (str), ..., claim_stance (str)]}
- IMPORTANT: The length of the list must be exactly equal to the number of claims.


TASK 3: Argument Relations Identification and Classification
- Identify relationships between arguments by determining which arguments support or attack other arguments.
- For each related argument pair, classify the relationship as either: "Support"(S) or "Attack"(A).
- IMPORTANT: Only arguments that appear in the same line in the essay can be related.
- You must return a JSON in the following format: {"relation_types": {'Paragraph 1': [(target_index (int), source_index (int), relation_type (str)), (target_index (int), source_index (int), relation_type (str)), ...], 'Paragraph 2': [(target_index (int), source_index (int), relation_type (str)), (target_index (int), source_index (int), relation_type (str)), ...], ...}}
- Note: Indices are 1-based, referring to the position in the provided arguments list.

Example:

### Output:
{"argument_types": ['M', 'C', 'P', 'P', 'P', 'P', 'C', 'P', 'P', 'C', 'M']}
{"claim_stances: ['S', 'A', 'S']}
{"relation_types": {'Paragraph 1': [], 'Paragraph 2': [(3, 2, 'S'), (4, 2, 'S'), (5, 2, 'S')], 'Paragraph 3': [(6, 7, 'S'), (8, 10, 'S'), (9, 10, 'S')], 'Paragraph 4': []}}

"""
```

### Build data files

```python
df_train = pe_df[pe_df.split == 'TRAIN'].reset_index()

data_file_train = []

for _, row in df_train.iterrows():

    question = row.usr_text
    answer = row.output
    
    data_file_train.append(formatting_fct(instruction, question, answer))
```
