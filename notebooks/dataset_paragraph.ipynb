```python
import itertools
```

```python
para_id = [0, 1, 2, 3, 4]
ac_count = [0, 1, 4, 5, 1]
```

```python
def generate_pairs(count):
    """Generate all pairs of (i, j) where i != j and neither are 0."""
    return [(i, j) for i, j in itertools.combinations(range(1, count), 2)]  # Starts from 1 to exclude 0

# List to store all paragraph pairs
all_pairs = []

# Variable to keep track of the cumulative count of pairs
cumulative_count = 1  # Start from 1 for the first paragraph

# Iterate through paragraphs
for i in range(len(para_id)):
    # Update the cumulative count by adding the current ac_count
    cumulative_count += ac_count[i]
    
    # Generate pairs based on the cumulative count of previous paragraphs, excluding 0
    pairs = generate_pairs(cumulative_count)
    
    # Add the pairs for this paragraph to the all_pairs list
    all_pairs.append(pairs)
```

```python
all_pairs
```

```python
def get_all_pairs(row):
    
    para_nr_acs = row.ac_count
    all_pairs = [(i, j) for i in range(1, para_nr_acs + 1) for j in range(1, para_nr_acs + 1) if i != j]
    
    return all_pairs
```

```python
paras_df['all_para_pairs'] = paras_df.apply(lambda row: get_all_pairs(row), axis=1)
```

```python
def offset_pair_indices(row):
    
    ac_count = row.ac_count
    p_ids = row.p_id
    para_pairs = row.all_para_pairs
    
    new_para_pairs = []
    
    for p in p_ids:
        
        offset = sum([elem for elem in ac_count[:p]])
        #para_pairs_p = para_pairs[p]
        new_para = [(i + offset, j + offset) for i, j in para_pairs[p]]
        new_para_pairs.append(new_para)
        
    return new_para_pairs
```

```python
essays_df["correct_para_pairs"] = essays_df.apply(lambda row: offset_pair_indices(row), axis=1)
```

```python
def offset_ar_indices(row):
    
    ac_count = row.ac_count
    p_ids = row.p_id
    para_pairs = row.AR_pairs
    #print(para_pairs)
    
    new_para_pairs = []
    
    for p in p_ids:
        
        offset = sum([elem for elem in ac_count[:p]])
        #para_pairs_p = para_pairs[p]
        #print(para_pairs[p])
        new_para = [(i + offset +1, j + offset +1) for i, j in eval(para_pairs[p])]
        new_para_pairs.append(new_para)
        
    return new_para_pairs
```

```python
def get_sorted_acs(row):
    
    #arg_bounds = eval([int(eval(elem)) for elem in row.arg_start_bounds])
    #print(type(eval(row.arg_start_bounds)))
    arg_comps = [elem for elem in eval(row.argument_components)]
    arg_bounds = [int(elem) for elem in eval(row.arg_start_bounds)]
    #print(arg_bounds)
    sorted_indices = sorted(range(len(arg_bounds)), key=lambda k: arg_bounds[k])
    
    return [arg_comps[i] for i in sorted_indices]
```
