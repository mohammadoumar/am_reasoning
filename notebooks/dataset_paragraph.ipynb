```python
import itertools
```

```python
para_id = [0, 1, 2, 3, 4]
ac_count = [0, 1, 4, 5, 1]
```

```python
def generate_pairs(count):
    """Generate all pairs of (i, j) where i != j and neither are 0."""
    return [(i, j) for i, j in itertools.combinations(range(1, count), 2)]  # Starts from 1 to exclude 0

# List to store all paragraph pairs
all_pairs = []

# Variable to keep track of the cumulative count of pairs
cumulative_count = 1  # Start from 1 for the first paragraph

# Iterate through paragraphs
for i in range(len(para_id)):
    # Update the cumulative count by adding the current ac_count
    cumulative_count += ac_count[i]
    
    # Generate pairs based on the cumulative count of previous paragraphs, excluding 0
    pairs = generate_pairs(cumulative_count)
    
    # Add the pairs for this paragraph to the all_pairs list
    all_pairs.append(pairs)
```

```python
all_pairs
```

```python
def get_all_pairs(row):
    
    para_nr_acs = row.ac_count
    all_pairs = [(i, j) for i in range(1, para_nr_acs + 1) for j in range(1, para_nr_acs + 1) if i != j]
    
    return all_pairs
```

```python
paras_df['all_para_pairs'] = paras_df.apply(lambda row: get_all_pairs(row), axis=1)
```
